```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import re
import pandas as pd
from scipy.optimize import leastsq # Levenberg-Marquadt Algorithm #
```

#Measurement 1

```python
class CuD(object):
    def __init__(self, filename):
        # Data Parameters:
        self.data = pd.read_csv(filename, sep='\t', header=0)
        self.columnnames = ['Frequency', 'Mean Amplitude', 'Error Amplitude', 'Mean Phaseshift', 'Error  Phaseshift']
        
        # Fit Parameters:
        self.fitrange = [130., 170.] # min, max
        self.p = [1, 0, 1] # [hwhm, peak center, intensity] #
        
        # Plot parameters:
        self.savenames = ['01.pdf', '02.pdf', '02.pdf']
        self.dpi = 150
        self.figsize=(15,5)
        self.xr = 20
        # Data styles
        self.ma = '.'
        self.mcol = 'k'
        self.dms = 0
        self.dls = ''
        #Fit styles
        self.fcol = 'r'
        self.fls = '-'
        self.fls2 = '--'
        self.bcol = 'r'
        self.bma = '|'
        self.bms = 20
        self.rescol = 'k'
        self.resma = '.'
        self.resls = ''

        
        
        
    def remove_commas_re(self, subj):
        return re.sub('[' + re.escape(''.join(',')) + ']', '.', subj)

    def formatted_data(self, data):
        return[float(self.remove_commas_re(d)) for d in data]
    
    def lorentzian(self, x,p):
            numerator =  (1)
            denominator = np.sqrt(( (p[1])**2 - (x/(2*np.pi))**2 )**2 + (p[0]*(x/(2*np.pi)))**2)
            y = p[2]*(numerator/denominator)
            return y

    def residuals(self, p,y,x):
            err = y - self.lorentzian(x,p)
            return err
    def phi(self, x,p):
        return np.arctan2((p[0]*(x/(2*np.pi))) , (p[1]**2 - (x/(2*np.pi))**2)) 
        
    def fit_lor(self, data, p):
       
    
        x = data[0]
        y = data[1]
        
        # initial values #
        # p = [5.0,520.0,12e3]  # [hwhm, peak center, intensity] #

        # optimization # 
        pbest = leastsq(self.residuals,p,args=(y,x),full_output=1)
        best_parameters = pbest[0]

        # fit to data #
        # fit = lorentzian(x, best_parameters)
        
        pfit, pcov, infodict, errmsg, success = pbest

        if (len(y) > len(p)) and pcov is not None:
            res_sqrd = [self.residuals(pfit, y, xx)**2 for xx in x]
            dof = len(y)-len(p)
            s_sq = np.sum(res_sqrd) / dof # chi^2-red.
            pcov = pcov * s_sq
        else:
            pcov = inf
        print 's_sp =', s_sq
        error = [] 
        for i in range(len(pfit)):
            try:
              error.append( np.sqrt(np.absolute(pcov[i][i])))
            except:
              error.append( 0.00 )
        pfit_leastsq = pfit
        perr_leastsq = np.array(error)
        
        return best_parameters, perr_leastsq
    
    def find_nearest(self, array,value):
        idx = np.argmin(np.array([abs(arr-value) for arr in array]))
        return array[idx]

    def plot(self, fit=False):
        data = self.data
        columnnames = self.columnnames
        dpi = self.dpi
        figsize = self.figsize
        savename = self.savenames
        
        formatted_data = self.formatted_data
        
        freq = np.array(formatted_data(data[columnnames[0]]))
        MAmp = np.array(formatted_data(data[columnnames[1]]))
        eAmp = np.array(formatted_data(data[columnnames[2]]))
        MPha = np.array(formatted_data(data[columnnames[3]]))
        ePha = np.array(formatted_data(data[columnnames[4]]))
        
        names = [d for d in data]
        
        fig1, axes = plt.subplots(1,2)
        ax1, ax2 = axes
        fig1.set_size_inches(figsize)
        ax1.errorbar(freq, MAmp, yerr=eAmp, \
                     color=self.mcol, marker=self.ma, ms=self.dms, ls=self.dls, \
                    label='data')
        ax1.set_xlabel(names[0])
        ax1.set_ylabel(names[1])
        if fit:
            min_x, max_x = self.fitrange[0], self.fitrange[1]
            borders = (freq>min_x) & (freq<max_x)
            x_fit = freq[borders]
            y_fit = MAmp[borders]
            fit_data = [x_fit, y_fit]
            best_par, errors = self.fit_lor(fit_data, self.p)
            y_opt = self.lorentzian(freq, best_par)
            
            the_label = 'fit: $\\nu_0 = '  + str(round(2*np.pi*best_par[1],3)) + \
                        '$\n$\\gamma = '   + str(round(best_par[0],3)) + \
                            '$\n$A_0 = '   + str(round(best_par[2],6)) + '$'
            ax1.plot(freq, y_opt, \
                     color=self.fcol, ls=self.fls, \
                    label=the_label
                    )
            ax1.plot([x_fit[0], x_fit[-1]],[y_fit[0], y_fit[-1]], \
                     color=self.bcol, marker=self.bma, ms=self.bms, ls=self.dls)
            
            ax2.plot(x_fit, self.residuals(best_par, y_fit, x_fit), \
                    color=self.rescol, marker=self.resma, ls=self.resls)
        ax1.legend(loc='best')
        
        
        fig1.savefig(savename[0], dpi=dpi)
    
        fig2, ax2 = plt.subplots(1,1)
        fig2.set_size_inches(figsize)
        ax2.errorbar(freq, MPha, yerr=ePha, \
                     color=self.mcol, marker=self.ma, ms=self.dms, ls=self.dls, \
                    label='data')
        if fit:
            ax2.plot(freq, self.phi(freq, best_par), \
                    color=self.fcol, ls=self.fls,  \
                    label=the_label
                    )
                     
        ax2.set_xlabel(names[0])
        ax2.set_ylabel(names[3])
        ax2.legend(loc='best')
    
        fig2.savefig(savename[1], dpi=dpi)
        
        if fit:
            
            lower, upper = 2*np.pi*best_par[1]-self.xr, 2*np.pi*best_par[1]+self.xr
            
            xs = np.linspace(lower, upper, num=1000)
            Anom = np.array(self.lorentzian(xs, best_par))
            phis = self.phi(xs, best_par)
            Acos = self.lorentzian(xs, best_par) * np.cos(phis)
            Asin = self.lorentzian(xs, best_par) * np.sin(phis)
            
            Amax = Anom.max()
            x_mid = xs[np.where(Anom == Amax)]
            fwhm_l = self.find_nearest(Anom[xs < x_mid], Amax/np.sqrt(2))
            fwhm_h = self.find_nearest(Anom[xs > x_mid], Amax/np.sqrt(2))
            x_l, x_h = xs[Anom == fwhm_l], xs[Anom == fwhm_h]

            fig3, ax3 = plt.subplots(1,1)
            fig3.set_size_inches(figsize)
            ax3.plot(xs, Anom, \
                     color=self.mcol, ls=self.fls, \
                    label='fit $A$')
            ax3.plot(xs, -Acos, \
                     color=self.fcol, ls=self.fls2, \
                    label='fit $-A\\cdot\\operatorname{cos}(\\omega)$')
            ax3.plot(xs, Asin,  \
                     color=self.fcol, ls=self.fls2, \
                    label='fit $A\\cdot\\operatorname{sin}(\\omega)$')
            
            ax3.plot((x_l, x_h), (fwhm_l, fwhm_h), 'r|', ms=1000)
            ax3.annotate("$2\\delta\\nu$",
                xy=(x_h, 0), xycoords='data',
                xytext=(0.75, 0.25), textcoords='axes fraction',
                arrowprops=dict(arrowstyle="->",
                                facecolor='gray'
                                ),
                size=15
                )
            ax3.set_xlabel(names[0])
            ax3.set_ylabel(names[3])
            ax3.set_xlim(lower, upper)
            ax3.legend(loc='best')

            fig3.savefig(savename[2], dpi=dpi)
            
    
filename = '../../StephanJohannes/Measurements/Nachtmessung_140Hz_pm_50Hz'
plot_1 = CuD(filename)
plot_1.p = [3,140,0.00007]
plot_1.fitrange = [130, 170]
plot_1.xr = 5
plot_1.plot(fit=True)
```
